```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=np.arange(0,20,0.2)
x
```

```python
theta=np.array([2,3])

noise=3*np.random.randn(100)
noise
```

```python
y_ideal=theta[0]+theta[1]*x
y_real=theta[0]+theta[1]*x+noise
```

```python
plt.plot(x,y_ideal,color='orange')
plt.scatter(x,y_real)
plt.show()
```
